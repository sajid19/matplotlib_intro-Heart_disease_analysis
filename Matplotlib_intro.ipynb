## Matpotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [11,22,33,44]
# set up data
fig,ax= plt.subplots(figsize=(10,10)) #width,height
#plot data
ax.plot(x,y)
#Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")
#Save and show the figure
fig.savefig("images/sample-plot.png")
                     
```

## Two Option for Subplots


```python
# Option1
fig, ((ax1,ax2), (ax3,ax4))= plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(butter_price.keys(),butter_price.values());
ax4.hist(np.random.randn(1000));
```

```python
ax1.plot(x,x/2);
```

```python
# option 2
fig,ax = plt.subplots(nrows=2,
                        ncols=2,
                     figsize = (10,5))
# plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(butter_price.keys(),butter_price.values());
ax[1,1].hist(np.random.randn(1000));
```

## pyplot vs matplotlib OO method
* when plotting something quickly use pyplot method
* when plotting something advance use OO method

```python
over_50 = heart_disease[heart_disease["age"]> 50]
over_50.head()
```

```python
#pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            );
```

```python
## OO method mixed with pyplot 
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c= "target",
            ax = ax);
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize = (10,6))
# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y= over_50["chol"],
                     c= over_50["target"]
                    )
# Customize the plot
ax.set(title = "Heart diseases and Cholestrol Level",
      xlabel= "Age",
      ylabel="Cholestrol");
#add a legend
ax.legend(*scatter.legend_elements(),title = "Target");
# add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle ="--")
```

```python
over_50.head()
```

```python
#subplots of col.age thalach
fig,(ax0 ,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize = (10,10))
# Add data to axO
scatter = ax0.scatter(x=over_50["age"],
                    y= over_50["chol"],
                     c= over_50["target"])
#customize
ax0.set(title = "Heart diseases and Cholestrol Level",
      xlabel= "Age",
      ylabel="Cholestrol");
# legend
ax0.legend(*scatter.legend_elements(),title = "Title");
# mean line
ax0.axhline(over_50["chol"].mean(),
           linestyle ="--");
#for ax1 thalash

scatter = ax1.scatter(x=over_50["age"],
                    y= over_50["thalach"],
                     c= over_50["target"])

ax1.set(title = "Heart diseases and Max Heart Rate",
      xlabel= "Age",
      ylabel="Max Heart Rate");

ax1.legend(*scatter.legend_elements(),title = "Title");


ax1.axhline(over_50["thalach"].mean(),
           linestyle ="--");
# add a title to fig
fig.suptitle("Heart Disease Analysis",fontsize =19, fontweight ="bold")
```

## Discussion about Heart disease Analysis

```python
#Set the style
plt.style.use( 'seaborn-whitegrid')

#subplots of col.age thalach
fig,(ax0 ,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize = (10,10))
# Add data to axO
scatter = ax0.scatter(x=over_50["age"],
                    y= over_50["chol"],
                     c= over_50["target"],# change color
                     cmap="plasma")
#customize
ax0.set(title = "Heart diseases and Cholestrol Level",
      xlabel= "Age",

        ylabel="Cholestrol");
#change x axis limit
ax0.set_xlim(50,80)

# legend
ax0.legend(*scatter.legend_elements(),title = "Title");
# mean line
ax0.axhline(over_50["chol"].mean(),
           linestyle ="--");
#for ax1 thalash

scatter = ax1.scatter(x=over_50["age"],
                    y= over_50["thalach"],
                     c= over_50["target"],
                     cmap="winter")

ax1.set(title = "Heart diseases and Max Heart Rate",
      xlabel= "Age",
      ylabel="Max Heart Rate");
#limit
ax1.set_xlim(50,80)
ax1.set_ylim(60,200)

ax1.legend(*scatter.legend_elements(),title = "Title");


ax1.axhline(over_50["thalach"].mean(),
           linestyle ="--");
# add a title to fig
fig.suptitle("Heart Disease Analysis",fontsize =20, fontweight ="bold")
```
